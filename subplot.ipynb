```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,4*np.pi,1000)
y=np.sin(x)
```

```python
fig,ax=plt.subplots()
ax.plot(x,y)
fig.show()
```

```python
x=np.random.uniform(0,1,100)
y=np.random.uniform(0,1,100)
ax.scatter(x,y)
```

```python
x=np.arange(10)
y=np.random.uniform(1,10,10)
ax.bar(x,y)
```

```python
z=np.random.uniform(0,1,(8,8))
ax.imshow(z)
```

```python
z=np.random.uniform(0,1,(8,8))
ax.contourf(z)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot1
x1=np.array([1,2,3])
y=np.array([2,4,6])
plt.subplot(2,2,1)
plt.plot(x1,y,"o")
plt.title("relationship b/w x1 and y ",color='brown')
plt.grid(axis="x",color="green",linestyle="--")
plt.grid(axis="y",color="green",linestyle="--")

#plot2
x2=np.array([5,10,15])
y=np.array([30,20,10])
plt.subplot(2,2,2)
plt.plot(x2,y,"o")
plt.title("relationship b/w x2 and y ",color='brown')
plt.grid(axis="x",color="green",linestyle="--")
plt.grid(axis="y",color="green",linestyle="--")
#plot3
x3=np.array([1,2,3])
y=np.array([2,4,6])
plt.subplot(2,2,3)
plt.plot(x3,y,"o")
plt.title("relationship b/w x3 and y ",color='brown')
plt.grid(axis="x",color="green",linestyle="--")
plt.grid(axis="y",color="green",linestyle="--")
#plot4
x4=np.array([10,20,30])
y=np.array([5,10,15])
plt.subplot(2,2,4)
plt.plot(x4,y,"o")
plt.title("relationship b/w x4 and y ",color='brown')
plt.grid(axis="x",color="green",linestyle="--")
plt.grid(axis="y",color="green",linestyle="--")
```
